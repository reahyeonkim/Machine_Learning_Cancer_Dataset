### 데이터 준비 (특성 스케일링, 훈련/테스트 데이터 분리)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
cancer_scaled = sc.fit_transform(cancer.data)

X_train, X_test, y_train, y_test = train_test_split(cancer_scaled, cancer.target, test_size=0.3)
```

### 이진 분류기 선택 / 학습

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_preds = lr.predict(X_test)
lr_proba = lr.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

def evaluation(y_test, pred=None, pred_proba=None):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    roc_auc = roc_auc_score(y_test, pred_proba)
    print('오차 행렬')
    print(confusion)
    print('정확도 : {0:.4f} 정밀도 {1:.4f} 재현율 {2:.4f} F1 {3:.4f} AUC {4:.4f}\n'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
evaluation(y_test, lr_preds, lr_proba)
```

### 그리드 서치를 통해서 규제 강도를 tuning (l1, l2, alpha)


```python
from sklearn.model_selection import GridSearchCV

param = {
    'penalty': ['l1', 'l2'],
    'C': [0.01, 0.05, 0.1, 1, 5, 10]
}

lr_clf = LogisticRegression(solver='liblinear')
grid_clf = GridSearchCV(lr_clf, param_grid=param, scoring='accuracy', cv=5)
grid_clf.fit(X_train, y_train)
print('최적 하이퍼 파라미터 : {0}, 최적 평균 정확도 {1:.3f}'.format(grid_clf.best_params_, grid_clf.best_score_))
```

```python
best_model = grid_clf.best_estimator_

y_pred = best_model.predict(X_test)
y_pred_proba = best_model.predict_proba(X_test)[:, 1]
```

```python
evaluation(y_test, y_pred, y_pred_proba)
```

### 최종

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

def scale(method='None', X_train=None, X_test=None):
    if method == 'Standard':
        sc = StandardScaler()
        X_scaled_train = sc.fit_transform(X_train)
        X_scaled_test = sc.transform(X_test)
    elif method == 'MinMax':
        mm = MinMaxScaler()
        X_scaled_train = mm.fit_transform(X_train)
        X_scaled_test =mm.transform(X_test)
    elif method == 'Log':
        X_scaled_train = np.log1p(X_train)
        X_scaled_test = np.log1p(X_test)
    else:
        X_scaled_train = X_train
        X_scaled_test = X_test
    return X_scaled_train, X_scaled_test
```

```python
def Search(model, X_train, X_test, y_train, y_test, params):
    grid_clf = GridSearchCV(model, param_grid=params, scoring='accuracy', cv=5)
    grid_clf.fit(X_train, y_train)
    print('최적 하이퍼 파라미터 : {0}, 최적 평균 정확도 {1:.3f}'.format(grid_clf.best_params_, grid_clf.best_score_))
    best_model = grid_clf.best_estimator_
    y_pred = best_model.predict(X_test)
    y_pred_proba = best_model.predict_proba(X_test)[:, 1]
    evaluation(y_test, y_pred, y_pred_proba)
```

```python
scale_method = ['None', 'Standard', 'MinMax', 'Log']

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3)

params = {
    'penalty': ['l1', 'l2'],
    'C': [0.01, 0.05, 0.1, 1, 5, 10],
}

for s in scale_method:
    X_scaled_train, X_scaled_test = scale(s, X_train, X_test)
    print("Scaled Method:", s)
    lr_clf = LogisticRegression(solver='liblinear', max_iter=1000)
    Search(lr_clf, X_scaled_train, X_scaled_test, y_train, y_test, params)
```
